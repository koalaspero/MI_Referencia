```python
from sklearn.impute import SimpleImputer
import pandas as pd

feat = feat.dropna(axis=1, how='all')  # Remove columns with all missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(feat)
X = pd.DataFrame(X_imputed, columns=feat.columns)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
model = LogisticRegression()

# Create the RFE model and specify the number of features to select
num_features_to_select = 10
rfe = RFE(estimator=model, n_features_to_select=num_features_to_select)
X_rfe = rfe.fit_transform(X, y)

# Display the selected features
selected_features = X.columns[rfe.support_]
print("Selected Features:", selected_features)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

# Convert X_rfe to a DataFrame with the selected features
X_rfe_df = pd.DataFrame(X_rfe, columns=selected_features)

# Split data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_rfe_df, y, test_size=0.3, random_state=42)

# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
rf_classifier.fit(X_train, Y_train)

# Make predictions on the test data
Y_pred = rf_classifier.predict(X_test)

# Calculate accuracy and display as a percentage
accuracy = accuracy_score(Y_test, Y_pred) * 100
print("Accuracy: {:.2f}%".format(accuracy))

```

```python
import pandas as pd

# Get feature importances
feature_importances = rf_classifier.feature_importances_

# Create a DataFrame with feature names and their importances using X_rfe_df
feature_importance_df = pd.DataFrame({'Feature': X_rfe_df.columns, 'Importance': feature_importances})

# Sort the DataFrame by feature importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print or view the DataFrame
print(X_rfe_df)

```

```python
import matplotlib.pyplot as plt

# Get the first 10 features and their importances
top_10_features = feature_importance_df.head(10)

# Create a barplot
plt.figure(figsize=(10, 6))
plt.barh(top_10_features['Feature'], top_10_features['Importance'], color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Top 10 Features by Importance')
plt.gca().invert_yaxis()  # Invert the y-axis for better readability
plt.show()

```
